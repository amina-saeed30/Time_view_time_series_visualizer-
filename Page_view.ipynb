```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

<h1>Load the dataset with correct parsing</h1>

```python
df = pd.read_csv("fcc-forum-pageviews.csv", parse_dates=['date'], index_col='date')
```

<h1>Line plot of page views over time</h1>

```python
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['value'], label='Page Views', color='blue')
plt.title("Page Views Over Time")
plt.xlabel("Date")
plt.ylabel("Page Views")
plt.legend()
plt.grid()
plt.show()
```

<h1>Rolling average to smooth trends</h1>

```python
df['Rolling Mean'] = df['value'].rolling(window=30).mean()

plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Rolling Mean'], label='30-Day Rolling Average', color='red')
plt.title("30-Day Rolling Average of Page Views")
plt.xlabel("Date")
plt.ylabel("Page Views")
plt.legend()
plt.grid()
plt.show()
```

<h1>Monthly average page views</h1>

```python
df['Month'] = df.index.to_period('M')
monthly_avg = df.groupby('Month')['value'].mean()

plt.figure(figsize=(12, 6))
sns.barplot(x=monthly_avg.index.astype(str), y=monthly_avg.values, color='green')
plt.xticks(rotation=45)
plt.title("Monthly Average Page Views")
plt.xlabel("Month")
plt.ylabel("Average Page Views")
plt.show()
```
